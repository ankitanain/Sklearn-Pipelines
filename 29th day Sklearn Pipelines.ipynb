# Titanic-without-using-pipelines

```python
import pandas as pd
import numpy as np

from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import accuracy_score

from sklearn.tree import DecisionTreeClassifier
```

```python
df=pd.read_csv('train.csv')
```

```python
df
```

```python
df.drop(columns=['PassengerId','Name','Ticket','Cabin'],inplace=True)

# these columns are not of much use so we are droping them
```

```python
df.head()
```

```python
X_train,X_test,y_train,y_test = train_test_split(df.drop('Survived',axis=1),df['Survived'],test_size=0.2,random_state=42)


# now we are doing train teast split, because we are going to perform decision tree so we can take all columns, i mean
# no need to take only 2 column as in logistic regression, becoz this is not linear model
```

```python
X_train.head(2)
```

```python
y_train.head()
```

```python
df.isnull().sum()
```

```python
# Applying imputation

si_age = SimpleImputer()
si_embarked = SimpleImputer(strategy='most_frequent')

X_train_age = si_age.fit_transform(X_train[['Age']])
X_train_embarked = si_embarked.fit_transform(X_train[['Embarked']])

X_test_age = si_age.transform(X_test[['Age']])
X_test_embarked = si_embarked.transform(X_test[['Embarked']])


# we are filling missing values using simple imputer, we are filling missing values of age using their mean values, but embarked
# is categorical data, so we fill them with their most frequent values
```

```python
X_train_embarked
```

```python
# one hot encoding Sex and Embarked

ohe_sex = OneHotEncoder(sparse_output=False,handle_unknown='ignore')
ohe_embarked = OneHotEncoder(sparse_output=False,handle_unknown='ignore')

X_train_sex = ohe_sex.fit_transform(X_train[['Sex']])
X_train_embarked = ohe_embarked.fit_transform(X_train_embarked)

X_test_sex = ohe_sex.transform(X_test[['Sex']])
X_test_embarked = ohe_embarked.transform(X_test_embarked)

# so we have sex & embarked as categorical data , so to perform algo on it firstly we have to convert them into numerical data
# handle ignore means in future when any unknown column come into existence then it will ignore them
```

```python
X_train_embarked
```

```python
X_train.head(2)
```

```python
X_train_rem = X_train.drop(columns=['Sex','Age','Embarked'])

# we have changed sex & embarked & age column of X_train data, so we have to drop these columns from X_train & add new formed
```

```python
# same for test data
X_test_rem = X_test.drop(columns=['Sex','Age','Embarked'])
```

```python
X_train_transformed = np.concatenate((X_train_rem,X_train_age,X_train_sex,X_train_embarked),axis=1)
X_test_transformed = np.concatenate((X_test_rem,X_test_age,X_test_sex,X_test_embarked),axis=1)

# we have concated remaing columns(except sex,age,embarked) and new columns like new age,sex,embarked
```

```python
X_test_transformed.shape
```

```python
clf = DecisionTreeClassifier()
clf.fit(X_train_transformed,y_train)
```

```python
y_pred = clf.predict(X_test_transformed)
y_pred
```

```python
accuracy_score(y_test,y_pred)*100
```

# Titanic-using-pipeline

```python
df1=pd.read_csv('train.csv')
```

```python
df1
```

### Let's Plan

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline,make_pipeline
from sklearn.feature_selection import SelectKBest,chi2
```

```python
df1.drop(columns=['PassengerId','Name','Ticket','Cabin'],inplace=True)
```

```python
# Step 1 -> train/test/split
X_train,X_test,y_train,y_test = train_test_split(df1.drop(columns=['Survived']),df1['Survived'],test_size=0.2,random_state=42)
```

```python
X_train.head()
```

```python
y_train.sample(5)
```

```python
# imputation transformer
trf1 = ColumnTransformer([
    ('impute_age',SimpleImputer(),[2]),
    ('impute_embarked',SimpleImputer(strategy='most_frequent'),[6])
],remainder='passthrough')
```

```python
# one hot encoding
trf2 = ColumnTransformer([
    ('ohe_sex_embarked',OneHotEncoder(sparse_output=False,handle_unknown='ignore'),[1,6])
],remainder='passthrough')
```

```python
# Scaling
trf3 = ColumnTransformer([
    ('scale',MinMaxScaler(),slice(0,10))
])
```

```python
# Feature selection
trf4 = SelectKBest(score_func=chi2,k=8)
```

```python
# train the model
trf5 = DecisionTreeClassifier()
```

## Create Pipeline

```python
pipe = Pipeline([
    ('trf1',trf1),
    ('trf2',trf2),
    ('trf3',trf3),
    ('trf4',trf4),
    ('trf5',trf5)
])
```

## Pipeline vs make_pipeline
Pipeline requires naming of steps, make_pipeline does not.

(Same applies to ColumnTransformer vs make_column_transformer)

```python
# Alternate Syntax
pipe = make_pipeline(trf1,trf2,trf3,trf4,trf5)
```

```python
# train
pipe.fit(X_train,y_train)
```

## Explore the pipeline

```python
# Code here
pipe.named_steps
```

```python
pipe.named_steps['trf1'].transformers_[0][1].statistics_
```

```python
# Display Pipeline

from sklearn import set_config
set_config(display='diagram')
```

```python
# Predict
y_pred = pipe.predict(X_test)
```

```python
y_pred
```

```python
accuracy_score(y_test,y_pred)*100
```

## cross validation using-pipeline

```python
# cross validation using cross_val_score
from sklearn.model_selection import cross_val_score
cross_val_score(pipe, X_train, y_train, cv=5, scoring='accuracy').mean()
```

## Grid search using pipeline

```python
# gridsearchcv
params = {
    'trf5__max_depth':[1,2,3,4,5,None]
}
```

```python
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(pipe, params, cv=5, scoring='accuracy')
grid.fit(X_train, y_train)
```

```python
grid.best_score_
```

```python
grid.best_params_
```
